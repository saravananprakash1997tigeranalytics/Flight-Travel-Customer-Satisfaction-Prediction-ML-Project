### Treating the Outliers in Age, Arrival Delay in Minutes and Departure Delay in Minutes Features

```python
df.loc[(df['Age']>76) | (df['Age']< 15), 'Age']=df.Age.value_counts().index[0]
```

```python
for i in ['Arrival Delay in Minutes', 'Departure Delay in Minutes']:
    q3=df[i].quantile(0.75)
    q1=df[i].quantile(0.25)
    IQR=q3-q1
    UL=q3+1.5*IQR
    LL=q1-1.5*IQR
    df.loc[df[i]>UL,i]=UL
    df.loc[df[i]<LL,i]=LL
```

### Treating Null Values (Here we are dropping the  null value rows bcoz the number of null values are very less compared to the entire dataset)

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('encoder',OneHotEncoder(),[0,1,3])],remainder='passthrough')
X = np.array(ct.fit_transform(X),dtype=np.float)
```

```python
X.shape
```

```python
X
```

### Model Selection

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, recall_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.70, random_state=40)
```

### Scaling by Standardization

```python
sc_X  = StandardScaler()
X_train_sc = sc_X.fit_transform(X_train)
X_test_sc = sc_X.transform(X_test)
```

### Scaling by Normalization

```python
min_max_scaler = MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(X_train)
X_test_minmax = min_max_scaler.fit_transform(X_test)
```

### Model Selection

### Logistic Regression Standardized Input Features

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
log_reg1=LogisticRegression(max_iter=2500)
log_reg1.fit(X_train_sc,y_train)
pred_log1=log_reg1.predict(X_test_sc)
metrics=classification_report(pred_log1,y_test)
```

```python
print('Confusion Matrix is\n',confusion_matrix(y_test,pred_log1))
print('Accuracy is', accuracy_score(y_test,pred_log1))
print(metrics)
```

### Logistic Regression Normalized Input Features

```python
log_reg2=LogisticRegression(max_iter=2500)
log_reg2.fit(X_train_minmax,y_train)
pred_log2=log_reg2.predict(X_test_minmax)
metrics=classification_report(pred_log2,y_test)
```

```python
print('Confusion Matrix is\n',confusion_matrix(y_test,pred_log2))
print('Accuracy is', accuracy_score(y_test,pred_log2))
print(metrics)
```




### K Nearest Neighbour

### Standardardized Input Features

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=359,metric='minkowski',p=2) 
knn.fit(X_train_sc,y_train)
pred_knn = knn.predict(X_test_sc)
metrics=classification_report(pred_knn,y_test)
```

```python
print('The K Value is Choosen as square root of number of samples')
print('Confusion Matrix is ',confusion_matrix(y_test,pred_knn))
print('Accuracy is', accuracy_score(y_test,pred_knn))
print('\n')
print(metrics)
```

### Normalized Input Featues

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=359,metric='minkowski',p=2) 
knn.fit(X_train_minmax,y_train)
pred_knn = knn.predict(X_test_minmax)
metrics=classification_report(pred_knn,y_test)
```

```python
print('The K Value is Choosen as square root of number of samples')
print('Confusion Matrix is ',confusion_matrix(y_test,pred_knn))
print('Accuracy is', accuracy_score(y_test,pred_knn))
print('\n')
print(metrics)
```

### Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
for d in range(20,30):
    dtc = DecisionTreeClassifier(criterion='entropy', max_depth=d,max_leaf_nodes=1000)
    dtc.fit(X_train,y_train)
    pred_dtc=dtc.predict(X_test)
    print("d=",d)
    print(accuracy_score(y_test,pred_dtc))
    metrics=classification_report(pred_dtc,y_test)
    print(metrics)
    
```

### In our proble to classify customers as satisfied or dissatisfied, best accuracy was achieved using Decision Tree Algorithm with maximum accuracy of 94%.
